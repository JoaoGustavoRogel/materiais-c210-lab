### Criação dos dados

```
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=0.5, shuffle=True)
```

### Visualização

```
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 150
plt.scatter(x[:, 0], x[:, 1], marker="o", edgecolors="black")
plt.show()
```

### Criação do KMeans

```
from sklearn.cluster import KMeans

modelo = KMeans(n_clusters=3)
modelo.fit(x)
```

### Predições

```
y_km = modelo.predict(x)

plt.scatter(x[y_km == 0, 0], x[y_km == 0, 1], c="green", marker="s", label="Grupo 1", edgecolors="black")
plt.scatter(x[y_km == 1, 0], x[y_km == 1, 1], c="blue", marker="o", label="Grupo 2", edgecolors="black")
plt.scatter(x[y_km == 2, 0], x[y_km == 2, 1], c="red", marker="v", label="Grupo 3", edgecolors="black")

plt.scatter(modelo.cluster_centers_[:, 0], modelo.cluster_centers_[:, 1], c="orange", label="Centros", marker="*")

plt.legend(scatterpoints=1)

plt.show()
```

### Scores

```
from sklearn.metrics import calinski_harabasz_score, silhouette_score

print("CH:", calinski_harabasz_score(x, modelo.labels_))
print("SC:", silhouette_score(x, modelo.labels_))
```
