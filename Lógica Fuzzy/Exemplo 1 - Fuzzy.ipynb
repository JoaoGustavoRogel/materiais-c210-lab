### Importando Pacotes

```
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

### Criação Antecedent

```
speed = ctrl.Antecedent(np.arange(0, 141, 1), "speed")

speed.automf(5)
speed.view()
```

### Criação Consequent

```
acceleration = ctrl.Consequent(np.arange(0, 101, 1), "acceleration")
acceleration.automf(5)
acceleration.view()
```

### Criação Regras

```
rule1 = ctrl.Rule(speed["poor"], acceleration["good"])
rule2 = ctrl.Rule(speed["mediocre"], acceleration["decent"])
rule3 = ctrl.Rule(speed["average"], acceleration["average"])
rule4 = ctrl.Rule(speed["decent"], acceleration["mediocre"])
rule5 = ctrl.Rule(speed["good"], acceleration["poor"])
```

### Criação do Sistema

```
acceleration_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
acceleration = ctrl.ControlSystemSimulation(acceleration_ctrl)
```

### Realizando Teste

```
acceleration.input["speed"] = 47
acceleration.compute()

ans = acceleration.output["acceleration"]
print("Aceleração de", ans, "%")
```
